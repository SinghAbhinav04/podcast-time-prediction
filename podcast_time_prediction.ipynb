# Plan for handling the null values

1)**find the name of the podcast and then avg it and  inplace of null replace it**

```python
def handle_null_val():
  for podcast in X_train['Podcast_Name'].unique():
    sum = X_train.loc[X_train['Podcast_Name']==podcast]['Episode_Length_minutes'].sum()
    count = X_train.loc[X_train['Podcast_Name']==podcast]['Episode_Length_minutes'].count()
    avg = sum/count
    X_train.loc[X_train['Podcast_Name']==podcast,'Episode_Length_minutes']=X_train.loc[X_train['Podcast_Name']==podcast,'Episode_Length_minutes'].fillna(avg)
```

```python
handle_null_val()
```

```python
X_train.info()
```

```python
def handle_null_val_2():
  for podcast in X_train['Podcast_Name'].unique():
    sum = X_train.loc[X_train['Podcast_Name']==podcast]['Guest_Popularity_percentage'].sum()
    count = X_train.loc[X_train['Podcast_Name']==podcast]['Guest_Popularity_percentage'].count()
    avg = sum/count
    X_train.loc[X_train['Podcast_Name']==podcast,'Guest_Popularity_percentage']=X_train.loc[X_train['Podcast_Name']==podcast,'Guest_Popularity_percentage'].fillna(avg)
```

```python
from sklearn.preprocessing import LabelEncoder , OneHotEncoder
```

```python
#label encoding for all obj ft except sentiment
label_encoder=LabelEncoder()
for col in X_train.columns:
  if X_train[col].dtype=="object" and col!="Episode_Sentiment":
    X_train[col]=label_encoder.fit_transform(X_train[col])
    X_test[col]=label_encoder.transform(X_test[col])
```

```python
X_train["Episode_Sentiment"]=label_encoder.fit_transform(X_train['Episode_Sentiment'])
X_test["Episode_Sentiment"]=label_encoder.transform(X_test['Episode_Sentiment'])
```

#Hyperparam tuning + RandomizedSearchCV

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
rand_params={
    "max_depth":[5,8,10],
    "min_samples_split":[5,7,9],
    "criterion":["squared_error"],
    "min_samples_leaf":[5,8,10],
}
```

```python
random_search = RandomizedSearchCV(
    estimator=dtr,
    param_distributions=rand_params,
    n_iter=5,
    cv=5,
    verbose=2,
    n_jobs=-1
)
random_search.fit(X_train, y_train)
```

```python
best_params = random_search.best_params_
best_params
```

```python
best_model = random_search.best_estimator_
best_model
```

```python
final_pred = best_model.predict(X_test)
final_pred
```

#Random Forest , XGBOOST , MLPR

```python
from sklearn.ensemble import RandomForestRegressor
import xgboost as xgb
```

```python
from sklearn.neural_network import MLPRegressor
```

```python
rfr = RandomForestRegressor(n_estimators=100, max_depth=10, random_state=42)
rfr.fit(X_train, y_train)
rfr_preds = rfr.predict(X_test)
rfr_preds
```

```python
xgbr = xgb.XGBRegressor(n_estimators=100, max_depth=6, learning_rate=0.1, random_state=42)
xgbr.fit(X_train, y_train)
xgbr_preds = xgbr.predict(X_test)
xgbr_preds
```

```python
mlpr = MLPRegressor(
        hidden_layer_sizes=(128, 64),
        activation='relu',
        solver='adam',
        alpha=0.001,
        learning_rate='adaptive',
        learning_rate_init=0.01,
        max_iter=500,
        early_stopping=True,
        random_state=42
    )
mlpr.fit(X_train , y_train)
mlpr_preds = mlpr.predict(X_test)
mlpr_preds
```

```python
num = 750000
submission = pd.DataFrame(columns=["id","Listening_Time_minutes"])
submission["id"]=range(num,num+len(mlpr_preds))
submission["Listening_Time_minutes"]=mlpr_preds
submission.head()
```
